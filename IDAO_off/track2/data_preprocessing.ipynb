# Sample experiments

```python
gb_f = pd.read_csv('../data/train.csv.zip')
```

```python
start_date = 62
```

```python
gb_f = train_set.groupby('user_id')
gb_f = gb_f.filter(lambda x: (x.id3.unique().size>2) & (x.shape[0]<350) & (x.date.diff(1).mean()+x.date.max()>=start_date))
```

```python
products = gb_f.id3.unique()
m = products.size
```

```python
users = gb_f.user_id.unique()
n = users.size
```

```python
n
```

```python
tfidf = np.empty((m,n))
```

```python
for i,user in enumerate(users):
    tfidf[:,i] = np.isin(products, gb_f[gb_f.user_id==user].id3)
```

```python
ptfidf = pd.DataFrame(tfidf, index=products, dtype=np.int8)
```

```python
n_components = 300
ipca = IncrementalPCA(n_components=300) #, batch_size=30)
```

```python
tfidf_ipca = ipca.fit_transform(tfidf)
```

```python
ptfidf_ipca = pd.DataFrame(tfidf_ipca, index=products)
```

```python
ptfidf_ipca.to_csv('../data/pcaed_products.csv')
```

```python
ptfidf_ipca = pd.read_csv('pcaed_products.csv', index_col=0)
```

```python
gb_f = gb_f.groupby('user_id')
gb_f = gb_f.filter(lambda x: (x.id3.unique().size>8) & (x.shape[0]<300) & (x.date.max()>=start_date - 7) )
```

```python
rdm_5k_usrs = np.random.choice(gb_f.user_id.unique(),53979, replace=False)

sbm = np.empty((53979,6), dtype=int)

for i,usr in enumerate(rdm_5k_usrs):
    usr_watches = gb_f[gb_f.user_id==usr]['id3']
    crossed_watches = gb_f[(gb_f.user_id==usr) & (gb_f.date>=start_date-21)]['id3']
    mean_usr_vec = ptfidf_ipca.loc[usr_watches].mean(axis=0)
    best_matches = ptfidf_ipca.dot(mean_usr_vec).sort_values(ascending=False)
    sbm[i] = np.array([usr,]+list(best_matches[~best_matches.index.isin(crossed_watches.as_matrix())][:5].index))
```

```python
submission = pd.DataFrame(sbm,  columns=['user_id', 'id3_1', 'id3_2', 'id3_3', 'id3_4', 'id3_5'])
```

```python
submission.to_csv('submission.csv', index=False)
```
